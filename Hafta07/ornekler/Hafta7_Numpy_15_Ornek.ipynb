# Hafta 7: NumPy ve Dosya İşlemleri
Bu notebook, NumPy oturumlarında işlenen 15 best-practice örneğini içerir.

## 1) NumPy ile Basit Dizi Oluşturma

```
import numpy as np
arr = np.array([1,2,3,4])
arr
```

## 2) shape, ndim, dtype

```
a = np.array([[1,2],[3,4]])
a.shape, a.ndim, a.dtype
```

## 3) Broadcasting örneği

```
arr = np.array([1,2,3])
arr + 10
```

## 4) İndeksleme & Slicing

```
a = np.arange(10)
a[2:7:2]
```

## 5) reshape ve transpose

```
a = np.arange(6).reshape(2,3)
a.T
```

## 6) np.random temel kullanımlar

```
np.random.seed(0)
np.random.randint(1,10,5)
```

## 7) mean, std

```
a = np.array([1,2,3,4,5])
a.mean(), a.std()
```

## 8) np.where koşullu seçim

```
a = np.array([5,2,8,1])
np.where(a>4, 'Büyük', 'Küçük')
```

## 9) NaN yönetimi

```
b = np.array([1, np.nan, 3])
np.nanmean(b)
```

## 10) np.loadtxt ile dosya okuma (örnek sim)

```
# gerçek dosya gereken ortamda çalışır
# np.loadtxt('dosya.txt')
print('loadtxt örneği')
```

## 11) np.savetxt ile yazma (demo)

```
np.savetxt('/mnt/data/demo.txt', np.array([1,2,3]))
'Kaydedildi'
```

## 12) argsort ve sıralama

```
x = np.array([30,10,20])
idx = np.argsort(x)
x[idx]
```

## 13) unique ve counts

```
vals = np.array([1,2,1,3,2])
np.unique(vals, return_counts=True)
```

## 14) Cumulative sum / prod

```
a = np.array([1,2,3])
a.cumsum(), a.cumprod()
```

## 15) Diff ve percentile

```
a = np.array([10,20,30])
np.diff(a), np.percentile(a,50)
```
